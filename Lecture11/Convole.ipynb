```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.io as io
import numpy as np
```

```python
img=io.imread('../images/stones.jpeg')
io.imshow(img)
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
io.imshow(img,cmap='gray')
```

```python
kernel=np.asarray([[-1.,-1.,-1.],
                 [-1.,8.0,-1.],
                 [-1.,-1.,-1.]])
print kernel
```

```python
new_image=np.zeros((img.shape[0]-kernel.shape[0]+1,img.shape[1]-kernel.shape[1]+1))
```

```python
#convole
for ix in range(new_image.shape[0]):
    for iy in range(new_image.shape[1]):
        im_patch=img[ix:ix+kernel.shape[0],iy:iy+kernel.shape[1]]
        h_prod=im_patch*kernel
        new_image[ix,iy]=max(0,h_prod.sum()) #Relu - removes all the negative if some value comes negative then it replaces it with 0
```

```python
io.imshow(new_image,cmap='gray')
```

```python
plt.imshow(new_image,cmap='gray')
```

```python
#to convert all the floating point value into int
#uint8 is unsigned int 8bit
plt.imshow(new_image.astype(np.uint8),cmap='gray')
```

```python
#generators it generates values one by one in order to save space,saves 1 element at a time

def myrange(t):
    k=0
    while k<t:
        yield k
        k+=1
        
z= myrange(10)
#print z.next()
#it is like a linked lst you are an initial point then via this u go further
for ix in z:
    print ix
```
