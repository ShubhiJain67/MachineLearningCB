## Convolutions

Write a function to create a simple convolution function for a CNN

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.io as io
```

```python
im = io.imread('../images/stones.jpeg')
img = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)

print img.shape

plt.imshow(img, cmap='gray')
```

```python
kernel = np.array([[-1., -1., -1.],
                   [-1., 8.0, -1.],
                   [-1., -1., -1.]])
print kernel
```

```python
new_image = np.zeros((img.shape[0]-kernel.shape[0]+1, img.shape[1]-kernel.shape[1]+1))
```

```python
# Convolve
for ix in range(new_image.shape[0]):
    for iy in range(new_image.shape[1]):
        im_patch = img[ix:ix+kernel.shape[0], iy:iy+kernel.shape[1]]
        h_prod = im_patch * kernel
        new_image[ix, iy] = max(0, h_prod.sum())
```

```python
plt.imshow(new_image.astype(np.uint8), cmap='gray')
```
